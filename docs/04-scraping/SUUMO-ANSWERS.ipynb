```python
from playwright.async_api import async_playwright
import pandas as pd
from io import StringIO

# Open browser
playwright = await async_playwright().start()
browser = await playwright.chromium.launch(headless=False)
page = await browser.new_page()

# Start from the first page
await page.goto("https://suumo.jp/jj/chintai/ichiran/FR301FC005/?sc=13101&ar=030&bs=040&sd=1&ta=13&ts=1")
```

```python
from bs4 import BeautifulSoup
import pandas as pd

# Get the page content
html = await page.content()

# Parse the page content with BeautifulSoup
soup = BeautifulSoup(html, 'lxml')

properties = []

# Find all property listings
listings = soup.select('.js-property')

for listing in listings:
    name = listing.select_one('.property_inner-title a').text if listing.select_one('.property_inner-title a') else None
    rent = listing.select_one('.detailbox-property-point').text if listing.select_one('.detailbox-property-point') else None
    management_fee = listing.select_one('.detailbox-property--col1 div:nth-child(2)').text if listing.select_one('.detailbox-property--col1 div:nth-child(2)') else None
    deposit = listing.select_one('.detailbox-property--col2 div:nth-child(1)').text if listing.select_one('.detailbox-property--col2 div:nth-child(1)') else None
    key_money = listing.select_one('.detailbox-property--col2 div:nth-child(2)').text if listing.select_one('.detailbox-property--col2 div:nth-child(2)') else None
    layout = listing.select_one('.detailbox-property--col3 div:nth-child(1)').text if listing.select_one('.detailbox-property--col3 div:nth-child(1)') else None
    size = listing.select_one('.detailbox-property--col3 div:nth-child(2)').text if listing.select_one('.detailbox-property--col3 div:nth-child(2)') else None
    direction = listing.select_one('.detailbox-property--col3 div:nth-child(3)').text if listing.select_one('.detailbox-property--col3 div:nth-child(3)') else None
    age = listing.select_one('.detailbox-property--col4 div:nth-child(2)').text if listing.select_one('.detailbox-property--col4 div:nth-child(2)') else None
    address = listing.select_one('.detailbox-property--col4').text if listing.select_one('.detailbox-property--col4') else None
    station_info = listing.select_one('.detailnote-box').text if listing.select_one('.detailnote-box') else None
    agency_name = listing.select_one('.detailnote-box-item div:nth-child(1)').text if listing.select_one('.detailnote-box-item div:nth-child(1)') else None
    agency_phone = listing.select_one('.detailnote-box-item div:nth-child(2)').text if listing.select_one('.detailnote-box-item div:nth-child(2)') else None
    property_url = listing.select_one('.property_inner-title a')['href'] if listing.select_one('.property_inner-title a') else None

    properties.append({
        "Property Name": name,
        "Rent": rent,
        "Management Fee": management_fee,
        "Deposit": deposit,
        "Key Money": key_money,
        "Floor Plan": layout,
        "Area Size": size,
        "Direction": direction,
        "Building Age": age,
        "Address": address,
        "Station Info": station_info,
        "Agency Name": agency_name,
        "Agency Phone": agency_phone,
        "Property URL": f'https://suumo.jp{property_url}' if property_url else None
    })

# Convert to a DataFrame
df = pd.DataFrame(properties)
df.head()  # Display the first few rows to verify
```

```python
from bs4 import BeautifulSoup
import pandas as pd

# Initialize an empty list to hold all the scraped properties
all_properties = []

# Start scraping from page 1
page_number = 1
while True:
    # Navigate to the next page
    url = f"https://suumo.jp/jj/chintai/ichiran/FR301FC005/?sc=13101&ar=030&bs=040&sd=1&ta=13&ts=1&page={page_number}"
    await page.goto(url)
    
    # Get the page content
    html = await page.content()
    
    # Parse with BeautifulSoup
    soup = BeautifulSoup(html, 'lxml')
    
    # Scrape the listings on the current page
    listings = soup.select('.js-property')
    
    # Check if there are any listings, if not break the loop
    if not listings:
        print(f"No more listings found. Stopping at page {page_number}.")
        break

    for listing in listings:
        name = listing.select_one('.property_inner-title a').text if listing.select_one('.property_inner-title a') else None
        rent = listing.select_one('.detailbox-property-point').text if listing.select_one('.detailbox-property-point') else None
        management_fee = listing.select_one('.detailbox-property--col1 div:nth-child(2)').text if listing.select_one('.detailbox-property--col1 div:nth-child(2)') else None
        deposit = listing.select_one('.detailbox-property--col2 div:nth-child(1)').text if listing.select_one('.detailbox-property--col2 div:nth-child(1)') else None
        key_money = listing.select_one('.detailbox-property--col2 div:nth-child(2)').text if listing.select_one('.detailbox-property--col2 div:nth-child(2)') else None
        layout = listing.select_one('.detailbox-property--col3 div:nth-child(1)').text if listing.select_one('.detailbox-property--col3 div:nth-child(1)') else None
        size = listing.select_one('.detailbox-property--col3 div:nth-child(2)').text if listing.select_one('.detailbox-property--col3 div:nth-child(2)') else None
        direction = listing.select_one('.detailbox-property--col3 div:nth-child(3)').text if listing.select_one('.detailbox-property--col3 div:nth-child(3)') else None
        age = listing.select_one('.detailbox-property--col4 div:nth-child(2)').text if listing.select_one('.detailbox-property--col4 div:nth-child(2)') else None
        address = listing.select_one('.detailbox-property--col4').text if listing.select_one('.detailbox-property--col4') else None
        station_info = listing.select_one('.detailnote-box').text if listing.select_one('.detailnote-box') else None
        agency_name = listing.select_one('.detailnote-box-item div:nth-child(1)').text if listing.select_one('.detailnote-box-item div:nth-child(1)') else None
        agency_phone = listing.select_one('.detailnote-box-item div:nth-child(2)').text if listing.select_one('.detailnote-box-item div:nth-child(2)') else None
        property_url = listing.select_one('.property_inner-title a')['href'] if listing.select_one('.property_inner-title a') else None

        all_properties.append({
            "Property Name": name,
            "Rent": rent,
            "Management Fee": management_fee,
            "Deposit": deposit,
            "Key Money": key_money,
            "Floor Plan": layout,
            "Area Size": size,
            "Direction": direction,
            "Building Age": age,
            "Address": address,
            "Station Info": station_info,
            "Agency Name": agency_name,
            "Agency Phone": agency_phone,
            "Property URL": f'https://suumo.jp{property_url}' if property_url else None
        })

    # Check if there's a "Next" button to continue paginating
    next_button = soup.select_one('a:contains("次へ")')
    if not next_button:
        print(f"Reached the last page at {page_number}.")
        break

    # Move to the next page
    page_number += 1

# Convert the data to a DataFrame
df = pd.DataFrame(all_properties)
df.head()  # Display the first few rows to verify
```
