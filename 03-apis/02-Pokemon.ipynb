# Pokemon API

We'll now look at the [Pokeapi.co Pokemon API](https://pokeapi.co/). Documentation is available [here](https://pokeapi.co/docs/v2).

## Getting information about Snorlax

```python
import requests

response = requests.get("https://pokeapi.co/api/v2/pokemon/snorlax")
data = response.json()
data
```

```python
import pandas as pd
pd.options.display.max_columns = None

df = pd.json_normalize(data)
df.head()
```

```python
import pandas as pd

df = pd.json_normalize(data['moves'])
df.head()
```

```python
df = pd.json_normalize(data['abilities'])
df.head()
```

```python
df = pd.json_normalize(data['stats'])
df.head()
```

## Getting information about all pokemon

[This endpoint](https://pokeapi.co/docs/v2#pokemon) looks like it's about all pokemon.

```python
import requests

response = requests.get("https://pokeapi.co/api/v2/pokemon/?limit=20")
data = response.json()
data
```

```python
df = pd.json_normalize(data['results'])
df
```

But how do we add all of the extra columns? Maybe we can [ask ChatGPT if it has any ideas](https://chatgpt.com/share/66f1f2ca-e03c-800d-b2d2-a72cd483f8b7).

```python
# Function to retrieve and normalize Pokmon data from the URL
def get_pokemon_data(url):
    response = requests.get(url)
    return pd.json_normalize(response.json())

# Loop over each URL in the dataframe, collect the normalized data
pokemon_dataframes = []
for url in df['url']:
    pokemon_data = get_pokemon_data(url)
    pokemon_dataframes.append(pokemon_data)

# Concatenate all dataframes into one
complete_pokemon_df = pd.concat(pokemon_dataframes, ignore_index=True)

# Show the complete dataframe
complete_pokemon_df.head()
```
